```python
import pycaret
```

```python
import pandas as pd 
df = pd.read_csv('train.csv')
df
```

```python
from pycaret.classification import setup
```

```python
exp_clf101 = setup(data = df, target = 'Survived', session_id=123) 
```

```python
from pycaret.classification import *
rf = create_model('rf')
```

```python
best = compare_models()
```

```python
evaluate_model(best)
```

```python
tune_model(best)
```

```python
test = pd.read_csv('tested.csv')
test
```

```python
tuned_model = tune_model(best)
```

```python
plot_model(tuned_model, plot = 'confusion_matrix')
```

```python
evaluate_model(tuned_model)
```

```python
final_model = finalize_model(tuned_model)
```

```python
unseen_predictions = predict_model(final_model, data=test)
unseen_predictions.head()
```

```python
save_model(final_model,'Final gradient boosting Model')
```
